```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/home/ali-mirza/AI-ML/data/Titanic-Dataset.csv')
df.head()
```

```python
df = df.drop(['PassengerId','Ticket','Cabin'],axis=1)
```

```python
df.columns
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.dropna(subset=['Embarked'], inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ['Sex','Embarked']
encoded_df = df.copy()
for col in columns:
    le = LabelEncoder()
    encoded_df[col] = le.fit_transform(encoded_df[col])

encoded_df

```

```python
encoded_df = encoded_df.drop(columns=['Name'])
```

```python
X = encoded_df.drop(columns=['Survived'])
Y = encoded_df['Survived']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC

model_SVC = SVC(kernel='rbf')

model_SVC.fit(X_train,y_train)
```

```python
y_predict = model_SVC.predict(X_test)
y_predict
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

accuracy_score(y_test,y_predict)
```

```python
confusion_matrix(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```

```python
from sklearn.neighbors import KNeighborsClassifier

model_KNN = KNeighborsClassifier(n_neighbors=5)
model_KNN.fit(X_train,y_train)
```

```python
y_pred = model_KNN.predict(X_test)
y_pred
```

```python
acc = accuracy_score(y_test,y_pred)

print(f"Your Model is {round(acc*100,2)}% Accurate")
```

```python
# ------------------|
#                   |
# CROSS VALIDATION  |
#                   |
#-------------------|

from sklearn.model_selection import cross_val_score
X_scaled = scaler.fit_transform(X)
score = cross_val_score(model_KNN, X_scaled,Y,cv=5,scoring='accuracy')
score.mean()
```
