```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
X = df.drop('species', axis=1)
Y = df['species']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC

model_SVM = SVC(gamma='auto')

model_SVM.fit(X_train,y_train)
```

```python
model_SVM.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV

classifier = GridSearchCV(model_SVM,{
    'C':[1,10,20,30],
    'kernel':['linear','rbf'],
},cv=5,return_train_score=False)
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.cv_results_
```

```python
result = pd.DataFrame(classifier.cv_results_)
result
```

```python
result[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn.neighbors import KNeighborsClassifier

model_KNN = KNeighborsClassifier(n_neighbors=9)

params = {
    "n_neighbors": [3, 5, 7, 9, 11],
    "weights": ["uniform", "distance"],
    "p": [1, 2]
}

grid = GridSearchCV(model_KNN,params,cv=5,return_train_score=False)
```

```python
grid.fit(X_train,y_train)
```

```python
result_KNN = pd.DataFrame(grid.cv_results_)
result_KNN
```

```python
best_model = grid.best_estimator_
best_model
```

```python
result_KNN[['param_n_neighbors','param_weights','mean_test_score']]
```
