```python
df.drop(
    columns=[
        "EmployeeCount",
        "EmployeeNumber",
        "Over18",
        "StandardHours"
    ],
    inplace=True
)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

label_cols = ["Attrition", "Gender", "OverTime"]

for col in label_cols:
    df[col] = le.fit_transform(df[col])
```

```python
one_hot_cols = [
    "BusinessTravel",
    "Department",
    "EducationField",
    "JobRole",
    "MaritalStatus"
]

df = pd.get_dummies(
    df,
    columns=one_hot_cols,
    drop_first=True,
    dtype=int
)
```

```python
df.head()
```

```python
X = df.drop("Attrition", axis=1)
y = df["Attrition"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "AdaBoost": AdaBoostClassifier(random_state=42),
    "SVM": SVC(),
    "XGBoost": XGBClassifier(random_state=42)
}
```

```python
scaled_models = [
    "Logistic Regression",
    "KNN",
    "SVM"
]
```

```python
result = []
for name,model in models.items():
    if name in scaled_models:
        model.fit(X_train_scaled,y_train)
        prediction = model.predict(X_test_scaled)
        acc = accuracy_score(y_test,prediction)
        f1 = f1_score(y_test,prediction)
        result.append(
            {
                "name":name,
                "accuracy":acc,
                "f1 score":f1
            }
        )
    else:
        model.fit(X_train,y_train)
        prediction = model.predict(X_test)
        acc = accuracy_score(y_test,prediction)
        f1 = f1_score(y_test,prediction)
        result.append(
            {
                "name":name,
                "accuracy":acc,
                "f1 score":f1
            }
        )
```

```python
pd.DataFrame(result).sort_values(by='accuracy',ascending=False)
```

```python
from sklearn.model_selection import cross_val_score

cv_result = []

for name,model in models.items():
    if name in scaled_models:
        score = cross_val_score(model,X_train_scaled,y_train,cv=5,scoring='f1')
    else:
        score = cross_val_score(model,X_train,y_train,cv=5,scoring='f1')
    cv_result.append({
        "Model": name,
        "Mean F1": score.mean(),
        "Std": score.std()
    })
```

```python
pd.DataFrame(cv_result).sort_values(by='Mean F1',ascending=False)
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_lr = {
    "C": [0.001, 0.01, 0.1, 1, 10, 100],
    "penalty": ["l1", "l2"],
    "solver": ["liblinear", "saga"],
    "max_iter": [500, 1000, 2000],
    "class_weight": [None, "balanced"]
}

param_xgb = {
    "n_estimators": [50, 100, 200, 300],
    "max_depth": [3, 5, 7, 9],
    "learning_rate": [0.01, 0.05, 0.1, 0.2],
    "subsample": [0.6, 0.8, 1.0],
    "colsample_bytree": [0.6, 0.8, 1.0],
    "gamma": [0, 0.1, 0.3, 0.5],
    "min_child_weight": [1, 3, 5],
    "reg_alpha": [0, 0.1, 1],
    "reg_lambda": [1, 5, 10]
}
```

```python
lr_clr = RandomizedSearchCV(
    LogisticRegression(),
    param_distributions=param_lr,
    n_iter=20,
    cv=5,
    scoring="f1",
    random_state=42,
    n_jobs=-1
)
```

```python
xgboost_clr = RandomizedSearchCV(
    XGBClassifier(random_state=42),
    param_distributions=param_xgb,
    n_iter=50,
    cv=5,
    scoring="f1",
    random_state=42,
    n_jobs=-1
)
```

```python
lr_clr.fit(X_train_scaled,y_train)
```

```python
xgboost_clr.fit(X_train,y_train)
```

```python
result = pd.DataFrame(lr_clr.cv_results_)
result
```

```python
result = pd.DataFrame(xgboost_clr.cv_results_)
result
```

```python
best_lr = LogisticRegression(
    solver="saga",
    penalty="l1",
    C=1,
    max_iter=500,
    random_state=42
)
```

```python
best_xgb = XGBClassifier(
    subsample=0.8,
    reg_lambda=1,
    reg_alpha=1,
    n_estimators=200,
    min_child_weight=3,
    max_depth=3,
    learning_rate=0.1,
    colsample_bytree=0.8,
    random_state=42
)
```

```python
best_lr.fit(X_train_scaled, y_train)
best_xgb.fit(X_train, y_train)
```

```python
lr_pred = best_lr.predict(X_test_scaled)
xgb_pred = best_xgb.predict(X_test)
```

```python
lr_acc = accuracy_score(y_test,lr_pred)
xgb_acc = accuracy_score(y_test,xgb_pred)
```

```python
print("LogisticRegression Accuracy Score is : ",lr_acc)
print("XGB Accuracy Score is : ",xgb_acc)
```

```python
print(classification_report(y_test, lr_pred))
print(confusion_matrix(y_test, lr_pred))

print(" Regression Precision:", precision_score(y_test, lr_pred))
print(" Regression Recall:", recall_score(y_test, lr_pred))
print(" Regression F1:", f1_score(y_test, lr_pred))

print(classification_report(y_test, xgb_pred))
print(confusion_matrix(y_test, xgb_pred))

print("Xg Boost Precision:", precision_score(y_test, xgb_pred))
print("Xg Boost Recall:", recall_score(y_test, xgb_pred))
print("Xg Boost F1:", f1_score(y_test, xgb_pred))
```
